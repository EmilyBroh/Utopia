```python
one_std_dev = govxx_pop['pct_gap'].std()
two_std_dev = govxx_pop['pct_gap'].std() * 2
three_std_dev = govxx_pop['pct_gap'].std() * 3
four_std_dev = govxx_pop['pct_gap'].std() * 4
five_std_dev = govxx_pop['pct_gap'].std() * 5
```

```python
for index, row in govxx_pop.iterrows():
    gap = abs(row['pct_gap'])  # This makes negative gaps positive

    if gap >= five_std_dev:
        govxx_pop.loc[index, 'dev'] = 'greater_than_5'
    elif gap >= four_std_dev:
        govxx_pop.loc[index, 'dev'] = 'greater_than_4'
    elif gap >= three_std_dev:
        govxx_pop.loc[index, 'dev'] = 'greater_than_3'
    elif gap >= two_std_dev:
        govxx_pop.loc[index, 'dev'] = 'greater_than_2'
    elif gap >= one_std_dev:
        govxx_pop.loc[index, 'dev'] = 'greater_than_1'
    else:
        govxx_pop.loc[index, 'dev'] = 'within_1_std_dev'
```
